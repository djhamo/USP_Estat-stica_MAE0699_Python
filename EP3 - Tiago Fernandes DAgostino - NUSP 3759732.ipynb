```python
import numpy as np
import re
import math

NUSP = [3,7,5,9,7,3,2]

def h(val1, val2):
    _alfa = NUSP[6] #2
    _beta = NUSP[5] #3
    _gama = NUSP[4] #4
    _delta = NUSP[3] #9
    
    if(val1 == 0 and val2 == 0):
        return _alfa/10
    if(val1 == 0 and val2 == 1):
        return _beta/10
    if(val1 == 1 and val2 == 0):
        return _gama/10
    if(val1 == 1 and val2 == 1):
        return _delta/10
    return 0    

def simula(M):
    X = [0,1,1]
    t = 0
    for t in range(0, M+1):
        U = np.random.uniform(0.0, 1.0)
        _h = h(X[t-2], X[t-1])
        #print("t[{}] h = {} u = {}".format(t, _h, U))
        if (U < _h):
            X.append(0)
        else:        
            X.append(1)            
    return X

def conta_transicao(lista, ocorrencia):
    busca = '(?=' + ocorrencia + ')'
    vetor = ''.join(map(str, lista))
    saida = re.findall(busca, vetor)
    #print ("Busca {} , Vetor {}, Saida {}, Tam Saida {}".format(busca, vetor, saida, len(saida)))
    return len(saida)

def mi_chapeu(n , seq):
    return conta_transicao(X[2:n+3], seq)/(n + 1)

def mi(n):
    m = {}

    m.update({"0": mi_chapeu(n , "0"), "1": mi_chapeu(n , "1")})
    
    m.update({"00": mi_chapeu(n , "00"), "01": mi_chapeu(n , "01"), "10": mi_chapeu(n , "10"), "11": mi_chapeu(n , "11")})
    
    m.update({
        "000": mi_chapeu(n , "000"), "001": mi_chapeu(n , "001"), "010": mi_chapeu(n , "010"), "011": mi_chapeu(n , "011"),
        "100": mi_chapeu(n , "100"), "101": mi_chapeu(n , "101"), "110": mi_chapeu(n , "110"), "111": mi_chapeu(n , "111")})
    
    return m

def prob(n):
    p = {}
    
    n00 = conta_transicao(X[2:n+3], '00')
    n01 = conta_transicao(X[2:n+3], '01')
    n10 = conta_transicao(X[2:n+3], '10')
    n11 = conta_transicao(X[2:n+3], '11')
    
    p.update({"00": n00/(n00+n01), "01": n01/(n00+n01), "10": n10/(n10+n11), "11": n11/(n10+n11)})
    
    n000 = conta_transicao(X[2:n+3], '000')
    n001 = conta_transicao(X[2:n+3], "001")
    n010 = conta_transicao(X[2:n+3], "010")
    n011 = conta_transicao(X[2:n+3], "011")
    n100 = conta_transicao(X[2:n+3], "100")
    n101 = conta_transicao(X[2:n+3], "101")
    n110 = conta_transicao(X[2:n+3], "110")
    n111 = conta_transicao(X[2:n+3], "111")
    
    p.update({"000": n000/(n000+n001), "001": n001/(n000+n001), "010": n010/(n010+n011), "011": n011/(n010+n011)})
    p.update({"100": n100/(n100+n101), "101": n101/(n100+n101), "110": n110/(n110+n111), "111": n111/(n110+n111)})

    n0000 = conta_transicao(X[2:n+3], '0000')
    n0001 = conta_transicao(X[2:n+3], "0001")
    n0010 = conta_transicao(X[2:n+3], "0010")
    n0011 = conta_transicao(X[2:n+3], "0011")
    n0100 = conta_transicao(X[2:n+3], "0100")
    n0101 = conta_transicao(X[2:n+3], "0101")
    n0110 = conta_transicao(X[2:n+3], "0110")
    n0111 = conta_transicao(X[2:n+3], "0111")
    n1000 = conta_transicao(X[2:n+3], '1000')
    n1001 = conta_transicao(X[2:n+3], "1001")
    n1010 = conta_transicao(X[2:n+3], "1010")
    n1011 = conta_transicao(X[2:n+3], "1011")
    n1100 = conta_transicao(X[2:n+3], "1100")
    n1101 = conta_transicao(X[2:n+3], "1101")
    n1110 = conta_transicao(X[2:n+3], "1110")
    n1111 = conta_transicao(X[2:n+3], "1111")

    p.update({"0000": n0000/(n0000+n0001), "0001": n0001/(n0000+n0001), "0010": n0010/(n0010+n0011), "0011": n0011/(n0010+n0011)})
    p.update({"0100": n0100/(n0100+n0101), "0101": n0101/(n0100+n0101), "0110": n0110/(n0110+n0111), "0111": n0111/(n0110+n0111)})

    p.update({"1000": n1000/(n1000+n1001), "1001": n1001/(n1000+n1001), "1010": n1010/(n1010+n1011), "1011": n1011/(n1010+n1011)})
    p.update({"1100": n1100/(n1100+n0101), "1101": n1101/(n1100+n1101), "1110": n1110/(n1110+n1111), "1111": n1111/(n1110+n1111)})

    return p
```
